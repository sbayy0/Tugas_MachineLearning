# 1. Import Library

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

# 2. Load Dataset Iris

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

# 3. Split Data

```python
X = df[iris.feature_names]
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# 4. Buat Model Decision Tree

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

# 5. Prediksi dan Evaluasi

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

# 6. Visualisasi Decision Tree

```python
plt.figure(figsize=(12,8))
plot_tree(model, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

# 7. Kesimpulan

- Model berhasil mencapai akurasi 100% pada data uji, yang menunjukkan bahwa model sangat cocok untuk dataset ini.
- Hasil visualisasi menunjukkan bahwa fitur `petal length` adalah pemisah paling kuat antara kelas.
- Dataset Iris sangat cocok untuk digunakan sebagai studi kasus awal dalam pembelajaran supervised learning (klasifikasi).

Catatan: Akurasi 100% bisa jadi tanda overfitting pada dataset lain yang lebih kompleks, namun masih wajar untuk dataset ini karena fiturnya cukup terpisah dengan jelas.
